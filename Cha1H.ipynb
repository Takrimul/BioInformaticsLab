BA1H: Find All Approximate Occurrences of a Pattern in a String

```python
def hamming_distance(s1, s2):
    # Calculate the Hamming distance between two strings of equal length
    return sum(ch1 != ch2 for ch1, ch2 in zip(s1, s2))

def approximate_pattern_matching(Pattern, Text, d):
    positions = []  # Store starting positions where Pattern appears with at most d mismatches

    k = len(Pattern)
    for i in range(len(Text) - k + 1):
        substring = Text[i:i + k]
        if hamming_distance(Pattern, substring) <= d:
            positions.append(i)

    return positions

# Example usage:
with open(r"E:\Download_main\Downloads\rosalind_ba1h.txt") as file:
    Pattern = file.readline().strip()
    Text = file.readline().strip()
    d = int(file.readline().strip())
# Pattern = "ATTCTGGA"
# Text = "CGCCCGAATCCAGAACGCATTCCCATATTTCGGGACCACTGGCCTCCACGGTACGGACGTCAATCAAATGCCTAGCGGCTTGTGGTTTCTCCTACGCTCC"
# d = 3
result = approximate_pattern_matching(Pattern, Text, d)
print(" ".join(map(str, result)))  # Output: 6 7 26 27 78
```
