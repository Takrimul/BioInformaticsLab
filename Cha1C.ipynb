Chapter 1 problem statement : Find the Reverse Complement of a String

```python
def findReverseComplement(s):
    """
    Find the reverse complement of a DNA string.
    """
    return ''.join([{'A':'T', 'C':'G', 'G':'C', 'T':'A'}[B] for B in s[::-1]])
```
